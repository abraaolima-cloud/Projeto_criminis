```python
import pandas as pd
import matplotlib.pyplot as plt


print(pd.__version__)
```

```python
##analise primeiro semestre
df1 = pd.read_excel("SPDadosCriminais_2025.xlsx",'JAN_A_JUN_2025')
df1.head()
df1.info()
```

```python
#analise segundo semestre

df2 = pd.read_excel("SPDadosCriminais_2025.xlsx",'JUL_A_DEZ_2025')
df2.head()
df2.info()
```

```python
#JUNTANDO OS DOIS SEMESTRES#
df = pd.concat([df1, df2], ignore_index=True)
```

```python
#VERIFICANDO VALORES NULOS 
df.isnull().sum()
```

```python
#VERIFICANDO QUANTIDADE DE LINHAS#
df.shape
```

```python

teste= 12590/1161132 * 100
print (teste)
#PODEMOS VER QUE OS REGISTROS SEM BAIXOS CORRESPONDEM A 1 % DO TAMANHO AMOSTRAL, COMO MINHA ANALISE É ESPECIFICA VOU RETIRAR ESSES VALORES#
```

```python
#checando se o registro de bairros nulos está concentrado em um unico grupo#

df[df["BAIRRO"].isna()]["NATUREZA_APURADA"].value_counts(normalize=True)

```

```python
#EXCLUINDO REGISTROS COM BAIRROS VAZIOS#

df_sem_bairro_vazios = df.dropna(subset=["BAIRRO"])
df_sem_bairro_vazios.isnull().sum()
```

```python
#CHEGANDO SE HOUVE A LIMPEZA#
df.shape[0]
df_sem_bairro_vazios.shape[0]
```

```python
#RANCKING POR BAIRRO#

ranking_bairro = (
    df_sem_bairro_vazios.groupby("BAIRRO")
    .size()
    .reset_index(name="Total_Crimes")
    .sort_values("Total_Crimes", ascending=False)
)
```

```python
ranking_bairro.head(10)
```

```python
#DEIXANDO TUDO MAIUSCULO NO BAIRRO#

df_sem_bairro_vazios["BAIRRO"] = (
    df_sem_bairro_vazios["BAIRRO"]
    .str.upper()
    .str.strip()
)

```

```python
ranking_bairro = (
    df_sem_bairro_vazios
    .groupby("BAIRRO")
    .size()
    .reset_index(name="Total_Crimes")
    .sort_values("Total_Crimes", ascending=False)
)



```

```python
ranking_bairro.head(10)
```

```python
#DELIMITANDO A ANALISE POR BAIRRO URBANO# 

df_bairro_urbano = df_sem_bairro_vazios[df_sem_bairro_vazios["BAIRRO"] != "RURAL"]

#Foi realizada a padronização dos nomes dos bairros para evitar duplicidade de categorias causadas por variações de capitalização e espaços em branco.
# Além disso, categorias genéricas como "RURAL" foram tratadas separadamente, por não representarem bairros urbanos específicos.
```

```python
ranking_bairro = (
    df_bairro_urbano
    .groupby("BAIRRO")
    .size()
    .reset_index(name="Total_Crimes")
    .sort_values("Total_Crimes", ascending=False)
)
ranking_bairro.head(10)
```

```python
#PRECISO RELACIONAR O CENTRO COM O MUNICIPIO PARA TER MAIS INFORMAÇÕES, CENTRO DA ONDE?#
df_bairro_urbano["BAIRRO_MUNICIPIO"] = (
     df_bairro_urbano["BAIRRO"] + " - " + df_bairro_urbano["NOME_MUNICIPIO"]
)
```

```python
ranking_bairro = (
    df_bairro_urbano
    .groupby("BAIRRO_MUNICIPIO")
    .size()
    .reset_index(name="Total_Crimes")
    .sort_values("Total_Crimes", ascending=False)
)
ranking_bairro.head(10)

#OS 10 BAIRROS MAIS PERIGOSOS FICAM NA CIDADE DE SP#
```

```python
#VERIFICANDO MUDANÇAS#
df_bairro_urbano.head()
df_bairro_urbano.info()
```

```python
df_roubos_sp = df_bairro_urbano[
    df_bairro_urbano["NATUREZA_APURADA"] == "ROUBO - OUTROS"
]

ranking_bairro = (
    df_roubos_sp
    .groupby("BAIRRO_MUNICIPIO")
    .size()
    .reset_index(name="Total_Roubos")
    .sort_values("Total_Roubos", ascending=False)
)

ranking_bairro.head(20)
#OS 10 BAIRROS MAIS PERIGOSOS FICAM NA CIDADE DE SP#
```

```python
#tirando cidade de são paulo#

df_roubos_sp_sem_capital = df_roubos_sp[
    ~df_roubos_sp["NOME_MUNICIPIO"].str.upper().isin(["S.PAULO", "SÃO PAULO"])
]

ranking_municipios_sem_sp = (
    df_roubos_sp_sem_capital
    .groupby("BAIRRO_MUNICIPIO")
    .size()
    .reset_index(name="Total_Roubos")
    .sort_values("Total_Roubos", ascending=False)
)

ranking_municipios_sem_sp.head(15)



```
